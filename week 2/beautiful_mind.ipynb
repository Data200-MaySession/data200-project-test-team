### Importing necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Probability of success in a single attempt
p = 0.003
```

```python
# Number of attempts from 1 to 30
attempts = 10000
n_attempts = np.arange(1, attempts)

prob_of_failure = 1 - p

prob_of_failure_n_attempts = prob_of_failure ** n_attempts

# Probability of at least one success in n attempts
prob_at_least_one_success = 1 - (1 - p) ** n_attempts
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_attempts, prob_at_least_one_success, marker='o', linestyle='-', color='royalblue')
plt.title("Probability of At Least One Success vs Number of Attempts", fontsize=14)
plt.xlabel("Number of Attempts", fontsize=12)
plt.ylabel("Probability of At Least One Success", fontsize=12)
plt.grid(True)
plt.ylim(0, 1.05)
plt.axhline(1, color='gray', linestyle='--', linewidth=0.8)
# plt.xticks(range(1, attempts))
plt.tight_layout()
plt.show()
```

## Conclusion

### The plot shows that as the number of attempts increases, the probability of getting at least one success approaches 1.

### So consistent attempts increase the likelihood of success, even with a low probability of success in each attempt.
